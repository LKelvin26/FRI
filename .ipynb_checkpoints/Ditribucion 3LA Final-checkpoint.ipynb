```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt

NF = 3
Kg = basis(NF, 2)  # Estado base
Ka = basis(NF, 1)  # Estado metastable
Ke = basis(NF, 0)  # Estado excitado

# Estado inicial
psi0 = Kg 

# Parámetros del sistema
Omega = 3.5  # Frecuencia de Rabi
Delta = 0    # Desintonía
ga = 0.015 
gd = 0.05
g = 0.95
deltat = 50000
ntraj = 1

# Operadores de colapso para todas las transiciones relevantes
Ceg = np.sqrt(g) * (Kg * Ke.dag())  # Emisión de fotón detectable (excited to base)
Cea = np.sqrt(gd) * (Ka * Ke.dag())  # Transición de excitado a metastable
Cag = np.sqrt(ga) * (Kg * Ka.dag())  # Transición de metastable a base
collapse = [Ceg, Cea, Cag]

# Hamiltoniano
H = Delta * (Ke * Ke.dag()) + 0.5 * Omega * (Ke * Kg.dag() + Kg * Ke.dag())

# Lista de tiempos para la simulación
tlist = np.linspace(0, 900, deltat)

# Lista para almacenar tiempos de espera entre colapsos de tipo 0
wait_times = []

# Simulación Monte Carlo para medir tiempos de espera
for _ in range(ntraj):
    result = mcsolve(H, psi0, tlist, collapse, [], 1)
    traj = result.col_times[0]  # Extraer tiempos de colapso de esta trayectoria
    col_events = result.col_which[0]  # Identificar qué colapsos ocurrieron

    # Medir tiempos entre colapsos de tipo 0 (excited to base)
    last_time = None
    for i, col_time in enumerate(traj):
        if col_events[i] == 0:  # Solo registramos tiempos entre colapsos de tipo 0
            if last_time is not None:
                wait_times.append(col_time - last_time)
            last_time = col_time  # Actualizar el último tiempo de colapso de tipo 0

# Mostrar resultados
print("Tiempos de espera entre colapsos de tipo 0 (excitado a base):", wait_times)

# Graficar los tiempos
plt.hist(wait_times, bins=50, alpha=0.7, label="Tiempos de espera entre colapsos de tipo 0")
plt.xlabel("Tiempo")
plt.ylabel("Frecuencia")
plt.legend()
plt.title("Distribución de Tiempos de Espera entre Colapsos de Tipo 0")
plt.show()
```
