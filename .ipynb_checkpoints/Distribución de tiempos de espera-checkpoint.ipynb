# Programa de la solución de las trayectorias cuánticas para el sistema de 3 niveles.
## Distrubución de tiempos de espera

```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt

NF = 3
Kg = basis(NF, 2)
Ka = basis(NF, 1)
Ke = basis(NF, 0)

# Estado inicial
psi0 = Kg 

# Definición de operadores de colapso y Hamiltoniano
Aaa = Ka*Ka.dag()
Aee = Ke*Ke.dag()
Agg = Kg*Kg.dag()
Aeg = Ke*Kg.dag()
Aag = Ka*Kg.dag()
Aea = Ke*Ka.dag()
Aae = Ka*Ke.dag()
Aga = Kg*Ka.dag()
Age = Kg*Ke.dag()

Omega = 3.5  # Frecuencia de Rabi
Delta = 0    # Desintonía
ga = 0.015 
gd = 0.05
g = 0.95
deltat=10000
ntraj=5000
```

```python

# Operadores de colapso
Ceg = np.sqrt(g) * Age
Cea = np.sqrt(gd) * Aae
Cag = np.sqrt(ga) * Aga
collapse = [Ceg, Cea, Cag]

# Hamiltoniano
H = Delta * Aee + 0.5 * Omega * (Aeg + Age)

# Definir el tiempo de la simulación con alta resolución temporal
tlist = np.linspace(0, 450, deltat)  # Muy alta resolución temporal

# Resolver la ecuación usando Monte Carlo, solo 1 trayectoria
MC_result = mcsolve(H, psi0, tlist, collapse, [Aaa, Aee, Agg], ntraj)


```

```python


all_collapse_times = []
metastable_counts = 0  # Contador para tiempos de espera provenientes del estado metaestable

for traj in MC_result.col_times:
    waiting_time = 0
    state = "ground"  # Comienza en el estado base
    
    for colapso_time in traj:
        waiting_time += colapso_time
        
        if state == "ground":
            # Detectamos la transición al estado excitado
            state = "excited"
        
        elif state == "excited":
            # Detectamos el paso al estado metaestable
            state = "metastable"
        
        elif state == "metastable":
            # Detectamos la transición al estado base y guardamos el tiempo de espera
            all_collapse_times.append(waiting_time)
            metastable_counts += 1  # Incrementamos el contador
            waiting_time = 0  # Reiniciamos el tiempo de espera para el siguiente ciclo
            state = "ground"  # Reiniciamos el estado al base
            break  # Salimos después de completar el ciclo base → excitado → metaestable → base
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos los parámetros de la distribución
xi = 4.0  # Parámetro ξ (ajustable)
I_bar = 2.0  # Parámetro promedio de intensidad Ī (ajustable)

# Definimos el rango de tiempos de espera T
T = np.linspace(0, 5, 1000)

# Calculamos la distribución de tiempos de espera w(T)
w_T = xi * I_bar * np.exp(-xi * I_bar * T)

# Graficamos la distribución
plt.plot(T, w_T)
plt.xlabel("Tiempo de espera (T)")
plt.ylabel("Distribución de tiempos de espera $w(T)$")
plt.title("Distribución de tiempos de espera")
plt.legend()
plt.grid(True)
plt.show()
```
