```python
import cupy as cp
import matplotlib.pyplot as plt
from qutip import basis, sigmax, sigmaz
import scipy.linalg

# Obtener el Qobj y convertirlo a CuPy
Ke_qobj = basis(2, 1)  # Qobj
Ke_numpy = Ke_qobj.full()  # Convertir a NumPy
Ke = cp.asarray(Ke_numpy)  # Convertir a CuPy

Kg_qobj = basis(2, 0)  # Qobj
Kg_numpy = Kg_qobj.full()  # Convertir a NumPy
Kg = cp.asarray(Kg_numpy)  # Convertir a CuPy

# Estado inicial
psi0 = Kg

# Parámetros del sistema
Omega = 3
g = 0.95
Desintonia = 0
n_traj = 500
tlist = cp.linspace(0, 20, 1000)  # Array de CuPy

# Operador de salto
Ceg = cp.sqrt(g) * cp.dot(Kg, Ke.T.conj())

# Hamiltoniano efectivo
H = (Omega / 2) * cp.asarray(sigmax().full()) + (Desintonia / 2) * cp.asarray(sigmaz().full()) - 1j * (g / 2) * cp.dot(Ke, Ke.T.conj())

# Función para simular una trayectoria
def carmichael_trajectory(psi0, H, Ceg, tlist, g):
    psi_t = psi0
    states = []
    dt = tlist[1] - tlist[0]

    for t in tlist:
        Heff = H - 1j / 2 * cp.dot(Ceg.T.conj(), Ceg)
        Heff_numpy = cp.asnumpy(Heff)  # Convertir a NumPy
        dt_numpy = float(dt)  # Asegurar que dt sea un float
        exp_Heff = scipy.linalg.expm(-1j * Heff_numpy * dt_numpy)  # Exponencial en NumPy
        exp_Heff_cp = cp.asarray(exp_Heff)  # Convertir de vuelta a CuPy
        psi_t = exp_Heff_cp @ psi_t
        psi_t = psi_t / cp.linalg.norm(psi_t)

        Pe = g * dt * cp.real(cp.dot(psi_t.T.conj(), cp.dot(Ceg.T.conj(), Ceg) @ psi_t))

        if cp.random.rand() < Pe:
            psi_t = cp.dot(Ceg, psi_t)
            psi_t = psi_t / cp.linalg.norm(psi_t)

        states.append(psi_t)

    return states

# Simular múltiples trayectorias en la GPU
trajectories = []
for _ in range(n_traj):
    traj = carmichael_trajectory(psi0, H, Ceg, tlist, g)
    trajectories.append(traj)
```

```python
import cupy as cp

# Inicializar avg_rho como una lista de matrices CuPy en ceros
avg_rho = [cp.zeros((2, 2), dtype=cp.complex128) for _ in range(len(tlist))]

# Calcular el promedio de las trayectorias
for traj in trajectories:
    for i, psi in enumerate(traj):
        avg_rho[i] += cp.dot(psi, psi.T.conj())  # Sumar los operadores densidad

# Dividir por el número de trayectorias para obtener el promedio
avg_rho = [rho / n_traj for rho in avg_rho]

# Operador de proyección en el estado excitado |e⟩⟨e|
Ke = cp.asarray(basis(2, 1).full())  # Asegurar que Ke es un array de CuPy
Pe_op = cp.dot(Ke, Ke.T.conj())  # Operador de proyección en estado excitado

# Calcular la población del estado excitado
Pe_t = [cp.trace(cp.dot(Pe_op, rho)).real for rho in avg_rho]  # Extraer parte real
```

```python
# Convertir tlist y Pe_t a NumPy para graficar
tlist_np = cp.asnumpy(tlist)  # Convertir tlist a NumPy
Pe_t_np = cp.asnumpy(cp.array(Pe_t))  # Convertir Pe_t a NumPy

# Graficar la población del estado excitado
plt.plot(tlist_np, Pe_t_np, label="Población |e⟩")
plt.xlabel('Tiempo')
plt.ylabel('Población del estado excitado')
plt.title('Evolución de la población del estado excitado')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import cupy as cp
import matplotlib.pyplot as plt
from qutip import basis, sigmax, sigmaz
import scipy.linalg

# Obtener el Qobj y convertirlo a CuPy
Ke_qobj = basis(2, 1)  # Qobj
Ke_numpy = Ke_qobj.full()  # Convertir a NumPy
Ke = cp.asarray(Ke_numpy)  # Convertir a CuPy

Kg_qobj = basis(2, 0)  # Qobj
Kg_numpy = Kg_qobj.full()  # Convertir a NumPy
Kg = cp.asarray(Kg_numpy)  # Convertir a CuPy

# Estado inicial
psi0 = Kg

# Parámetros del sistema
Omega = 3
g = 0.95
Desintonia = 0
n_traj = 3
tlist = cp.linspace(0, 20, 1000)  # Array de CuPy

# Operador de salto
Ceg = cp.sqrt(g) * cp.dot(Kg, Ke.T.conj())

# Hamiltoniano efectivo
H = (Omega / 2) * cp.asarray(sigmax().full()) + (Desintonia / 2) * cp.asarray(sigmaz().full()) - 1j * (g / 2) * cp.dot(Ke, Ke.T.conj())

# Función para simular múltiples trayectorias en paralelo
def carmichael_trajectories_parallel(psi0, H, Ceg, tlist, g, n_traj):
    # Convertimos psi0 a la forma correcta para múltiples trayectorias
    psi_t = cp.tile(psi0, (n_traj, 1)).T  # Estado inicial para todas las trayectorias
    states = cp.zeros((len(tlist), psi0.shape[0], n_traj), dtype=cp.complex64)
    dt = tlist[1] - tlist[0]

    for i, t in enumerate(tlist):
        # Calculamos el Hamiltoniano efectivo
        Heff = H - 1j / 2 * cp.dot(Ceg.T.conj(), Ceg)
        Heff_numpy = cp.asnumpy(Heff)  # Convertir a NumPy
        dt_numpy = float(dt)  # Asegurar que dt sea un float
        exp_Heff = scipy.linalg.expm(-1j * Heff_numpy * dt_numpy)  # Exponencial en NumPy
        exp_Heff_cp = cp.asarray(exp_Heff)  # Convertir de vuelta a CuPy

        # Evolución de las trayectorias
        # Cambiamos para asegurarnos que psi_t tenga la forma correcta (2 x n_traj)
        psi_t = cp.dot(exp_Heff_cp, psi_t)
        psi_t = psi_t / cp.linalg.norm(psi_t, axis=0, keepdims=True)  # Normalización

        # Calcular las probabilidades de salto para todas las trayectorias
        Pe = g * dt * cp.real(cp.einsum('ij,ji->i', psi_t.T.conj(), cp.dot(Ceg.T.conj(), Ceg) @ psi_t))

        # Generar números aleatorios para los saltos cuánticos
        random_vals = cp.random.rand(n_traj)

        # Aplicar el salto cuántico donde corresponde
        jump_indices = random_vals < Pe
        psi_t[:, jump_indices] = cp.dot(Ceg, psi_t[:, jump_indices])
        psi_t[:, jump_indices] = psi_t[:, jump_indices] / cp.linalg.norm(psi_t[:, jump_indices], axis=0, keepdims=True)  # Normalización post-salto

        # Almacenar el estado actual de todas las trayectorias
        states[i] = psi_t

    return states

# Simular múltiples trayectorias en paralelo
trajectories = carmichael_trajectories_parallel(psi0, H, Ceg, tlist, g, n_traj)
```

```python
# Inicializar avg_rho como una lista de matrices CuPy en ceros
avg_rho = [cp.zeros((2, 2), dtype=cp.complex128) for _ in range(len(tlist))]

# Calcular el promedio de las trayectorias
for traj in trajectories:
    for i, psi in enumerate(traj):
        avg_rho[i] += cp.dot(psi, psi.T.conj())  # Sumar los operadores densidad

# Dividir por el número de trayectorias para obtener el promedio
avg_rho = [rho / n_traj for rho in avg_rho]

# Operador de proyección en el estado excitado |e⟩⟨e|
Ke = cp.asarray(basis(2, 1).full())  # Asegurar que Ke es un array de CuPy
Pe_op = cp.dot(Ke, Ke.T.conj())  # Operador de proyección en estado excitado

# Calcular la población del estado excitado
Pe_t = [cp.trace(cp.dot(Pe_op, rho)).real for rho in avg_rho]  # Extraer parte real
```

```python
# Convertir tlist y Pe_t a NumPy para graficar
tlist_np = cp.asnumpy(tlist)  # Convertir tlist a NumPy
Pe_t_np = cp.asnumpy(cp.array(Pe_t))  # Convertir Pe_t a NumPy

# Graficar la población del estado excitado
plt.plot(tlist_np, Pe_t_np, label="Población |e⟩")
plt.xlabel('Tiempo')
plt.ylabel('Población del estado excitado')
plt.title('Evolución de la población del estado excitado')
plt.legend()
plt.grid(True)
plt.show()
```
