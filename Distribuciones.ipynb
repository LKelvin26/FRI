## Distribuciones Discretas

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definir símbolos
n, mu = sp.symbols('n mu', real=True, positive=True)

# Definir la distribución de Poisson P(n)
P_n = (sp.exp(-mu) * mu**n) / sp.factorial(n)

# Calcular simbólicamente los primeros cuatro momentos
first_moment = sp.summation(n * P_n, (n, 0, sp.oo))
second_moment = sp.summation(n**2 * P_n, (n, 0, sp.oo))
third_moment = sp.summation(n**3 * P_n, (n, 0, sp.oo))
fourth_moment = sp.summation(n**4 * P_n, (n, 0, sp.oo))

# Simplificar los momentos
first_moment = sp.simplify(first_moment)
second_moment = sp.simplify(second_moment)
third_moment = sp.simplify(third_moment)
fourth_moment = sp.simplify(fourth_moment)

# Imprimir los momentos
print(f"Primer momento (media): {first_moment}")
print(f"Segundo momento: {second_moment}")
print(f"Tercer momento: {third_moment}")
print(f"Cuarto momento: {fourth_moment}")

# Generar la gráfica para un valor específico de mu
mu_value = 3  # Cambia este valor según sea necesario
n_values = np.arange(0, 20)
P_values = [float(P_n.subs({mu: mu_value, n: i})) for i in n_values]

# Graficar la distribución de Poisson
plt.figure(figsize=(10, 6))
plt.bar(n_values, P_values, color='blue', alpha=0.7, label=f'Poisson Distribution ($\\mu={mu_value}$)')
plt.title("Distribución de Poisson", fontsize=14)
plt.xlabel("n", fontsize=16)
plt.ylabel("P(n)", fontsize=16)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definir los símbolos
n, N, p = sp.symbols('n N p', real=True, positive=True)

# Definir la distribución binomial P(n)
P_n = sp.binomial(N, n) * (p**n) * ((1-p)**(N-n))

# Calcular simbólicamente los primeros cuatro momentos
first_moment = sp.summation(n * P_n, (n, 0, N))
second_moment = sp.summation(n**2 * P_n, (n, 0, N))
third_moment = sp.summation(n**3 * P_n, (n, 0, N))
fourth_moment = sp.summation(n**4 * P_n, (n, 0, N))

# Simplificar los momentos
first_moment = sp.simplify(first_moment)
second_moment = sp.simplify(second_moment)
third_moment = sp.simplify(third_moment)
fourth_moment = sp.simplify(fourth_moment)

# Mostrar los momentos en formato "bonito" con LaTeX
from sympy.printing.latex import latex

# Primer momento
display(Math("Primer\\ momento:\\ " + sp.latex(first_moment)))

# Segundo momento
display(Math("Segundo\\ momento:\\ " + sp.latex(second_moment)))

# Tercer momento
display(Math("Tercer\\ momento:\\ " + sp.latex(third_moment)))

# Cuarto momento
display(Math("Cuarto\\ momento:\\ " + sp.latex(fourth_moment)))


# Generar la gráfica para valores específicos de N y p
N_value = 10  # Número de ensayos
p_value = 0.5  # Probabilidad de éxito
n_values = np.arange(0, N_value + 1)
P_values = [float(P_n.subs({N: N_value, p: p_value, n: i})) for i in n_values]

# Graficar la distribución binomial
plt.figure(figsize=(10, 6))
plt.bar(n_values, P_values, color='green', alpha=0.7, label=f'Binomial Distribution ($N={N_value}$, $p={p_value}$)')
plt.title("Distribución Binomial", fontsize=14)
plt.xlabel("n", fontsize=12)
plt.ylabel("P(n)", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()

```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definir las constantes y símbolos
n, E, mu, k_B, T = sp.symbols('n E mu k_B T', real=True, positive=True)

# Definir la función de Bose-Einstein (P(n))
beta = 1 / (k_B * T)  # Inversa de la energía térmica
P_n = 1 / (sp.exp(beta * (E - mu)) - 1)

# Calcular simbólicamente los primeros cuatro momentos
first_moment = sp.summation(n * P_n, (n, 0, sp.oo))
second_moment = sp.summation(n**2 * P_n, (n, 0, sp.oo))
third_moment = sp.summation(n**3 * P_n, (n, 0, sp.oo))
fourth_moment = sp.summation(n**4 * P_n, (n, 0, sp.oo))

# Simplificar los momentos
first_moment = sp.simplify(first_moment)
second_moment = sp.simplify(second_moment)
third_moment = sp.simplify(third_moment)
fourth_moment = sp.simplify(fourth_moment)

# Mostrar las fórmulas en formato "bonito" con LaTeX
from IPython.display import display, Math

# Primer momento
display(Math("Primer\\ momento:\\ " + sp.latex(first_moment)))

# Segundo momento
display(Math("Segundo\\ momento:\\ " + sp.latex(second_moment)))

# Tercer momento
display(Math("Tercer\\ momento:\\ " + sp.latex(third_moment)))

# Cuarto momento
display(Math("Cuarto\\ momento:\\ " + sp.latex(fourth_moment)))

# Gráfica de la distribución para valores específicos
k_B_value = 1.38e-23  # Constante de Boltzmann (J/K)
T_value = 300  # Temperatura (K)
mu_value = 1e-21  # Potencial químico (J)
E_values = np.linspace(1e-21, 1e-20, 100)  # Energía (J)

# Evaluar la distribución numéricamente
P_values = [1 / (np.exp((E - mu_value) / (k_B_value * T_value)) - 1) for E in E_values]

# Graficar la distribución
plt.figure(figsize=(10, 6))
plt.plot(E_values, P_values, label="Distribución de Bose-Einstein")
plt.title("Distribución de Bose-Einstein", fontsize=14)
plt.xlabel("Energía (E)", fontsize=12)
plt.ylabel("P(n)", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
