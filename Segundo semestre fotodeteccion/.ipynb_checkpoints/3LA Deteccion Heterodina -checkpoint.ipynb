# Deteccion Heterodina
## Hamiltoniano del sistema 
\begin{equation}
    H_{\text{sys}} = \Delta \, \sigma_+ \sigma_- + \frac{1}{2} \Omega \, (\sigma_+ + \sigma_-) 
\end{equation} 
## Hamiltoniano de la deteccion heterodina
\begin{equation}
    H_Z = H_{\text{sys}} + i \left( \gamma \langle \bar{\sigma}_+ \rangle_c + \sqrt{\gamma }\eta_z Z(t) \right) \sigma_-
\end{equation} 

## Incremento de Weiner complejo 
\begin{align}
    \eta Z &= \frac{dZ}{dt} \\
    dZ &= e^{i \omega' t} \, dW = \frac{1}{\sqrt{2}} \left( dW_1 + i \, dW_2 \right)
\end{align}

## Forma alternativa de las ecuaciones
\begin{equation}
    I_{Zc}(t) = \gamma \langle \bar{\sigma}_+ \rangle_c + \sqrt{\gamma }\eta_z \, Z
\end{equation}

\begin{equation}
    \frac{d|\Psi_c\rangle}{dt} = H_Z |\Psi_c\rangle
\end{equation}

\begin{equation}
    \frac{d|\Psi_c\rangle}{dt} = -i \left( H_{\text{sys}} + I_{Zc}(t) \sigma_- \right) |\Psi_c\rangle
\end{equation} 

```python
# Deteccion heterodina
import numpy as np
import qutip as qt
import matplotlib.pyplot as plt

NF=3
Kg = qt.basis(NF,2)
Ka = qt.basis(NF,1)
Ke = qt.basis(NF,0)

# Estado base
psi0 = Kg 
# Parámetros
         # Tasa de detección
t_max = 30        # Tiempo de simulación
dt = 0.01            # Paso de tiempo
Omega = 3.5            # Frecuencia de Rabi
steps = int(t_max / dt)
Delta = 1          # Desintonía
num_traj = 5000     # Número de trayectorias

ga = 0.015 
gd = 0.05
g = 0.95

# Espacio de hilbert
Aaa = Ka*Ka.dag()
Aee = Ke*Ke.dag()
Agg = Kg*Kg.dag()
Aeg = Ke*Kg.dag()
Aag = Ka*Kg.dag()
Aea = Ke*Ka.dag()
Aae = Ka*Ke.dag()
Aga = Kg*Ka.dag()
Age = Kg*Ke.dag()

#Valores esperados
ket_a = Aaa #Definiendo los valores esperados con la diagonal principal
ket_e = Aee
ket_g = Agg

expect_values = [Aee,Agg,Aaa] #Los valores esperados se ponen en forma de lista, no se si importe su orden

Ceg = np.sqrt(g)*Age 
Cea = np.sqrt(gd)*Aae
Cag = np.sqrt(ga)*Aga
c_ops = [Ceg ,Cea, Cag]

#Eficiencia del detector
eta=1

# Operadores cuánticos
sm = Age    # σ-
sp = Age        # σ+
Pe = sp * sm          
 
# Hamiltoniano del sistema (Rabi) CORREGIDO
H_sys = Delta * Aee + 0.5 * Omega * (Age+Aeg) ##  [1]
times = np.linspace(0, t_max, steps)
```

```python
# Función para simular trayectorias con detección heterodina
def simulate_heterodyne():
    psi_t = Kg  # Estado inicial |g⟩
    pop_e = []  # Población de |e⟩
    pop_a = []  # Población de |a⟩ (shelving)
    photocurrent = []   


    for _ in range(steps):
        # Ruido complejo (dW1 + i*dW2)
        dW1 = np.random.normal(0, np.sqrt(dt))
        dW2 = np.random.normal(0, np.sqrt(dt))
        
        # Operadores de medida heterodina (2 canales)
        L1 = np.sqrt(eta * g / 2) * Age  # Canal en fase
        L2 = 1j * np.sqrt(eta * g / 2) * Age  # Canal en cuadratura
        
        # Fotocorrientes
        I1 = qt.expect(L1 + L1.dag(), psi_t) + dW1 / dt
        I2 = qt.expect(L2 + L2.dag(), psi_t) + dW2 / dt
        
        # Evolución determinista (H_eff = H_sys - (i/2) Σ L†L)
        H_eff = H_sys - (1j/2) * (L1.dag() * L1 + L2.dag() * L2)
        dpsi = (-1j * H_eff * psi_t) * dt
        
        # Términos estocásticos
        dpsi += L1 * psi_t * dW1 + L2 * psi_t * dW2
        
        # Aplicar decaimientos (Lindblad)
        for op in c_ops:
            dpsi += (op * psi_t * np.random.normal(0, np.sqrt(dt)))
        
        psi_t = (psi_t + dpsi).unit()
        pop_e.append(qt.expect(Aee, psi_t))
        pop_a.append(qt.expect(Aaa, psi_t))
        photocurrent.append(I1+1j*I2)
        
    
    return pop_e, pop_a, photocurrent

# Simulación de múltiples trayectorias
traj_pop_e = np.zeros((num_traj, steps))
traj_pop_a = np.zeros((num_traj, steps))

trajectories_I = np.zeros((num_traj, steps), dtype=np.complex128)

for k in range(num_traj):
    pop_e, pop_a, I_t = simulate_heterodyne()
    traj_pop_e[k, :] = pop_e
    traj_pop_a[k, :] = pop_a
    trajectories_I[k, :] = I_t
 
    print(f"Trayectoria {k+1}/{num_traj}", end="\r")

```

```python
# Resultados promediados
avg_pop_e = np.mean(traj_pop_e, axis=0)
avg_pop_a = np.mean(traj_pop_a, axis=0)
avg_I = np.mean(trajectories_I, axis=0, dtype=np.complex128)


# Guardar resultados en el formato deseado
results = {
    'avg_pop': avg_pop_e,  # Población promedio de |e⟩
    'avg_pop_shelving': avg_pop_a,  # Población promedio de |a⟩ (shelving)
    'avg_I': avg_I, # Fotocorriente compleja
    'trajectories_I': traj_I  # Trayectorias de fotocorriente
}
```

```python
# Gráfico de fotocorrientes promediadas
plt.figure(figsize=(12, 6))
times = np.linspace(0, t_max, steps)
plt.plot(times, results['avg_I'], label="Detección heterodina", linewidth=2)

# Estilos de texto
plt.xlabel("Tiempo", fontsize=14)
plt.ylabel("Fotocorriente promediada ⟨I(t)⟩", fontsize=14)
plt.title("Fotocorriente Heterodina", fontsize=16)

# Pie de figura con parámetros
plt.figtext(0.5, -0.1, f"Ω={Omega}, Δ={Delta}, γ={g}, t_max={t_max}, dt={dt}, Trayectorias={num_traj}",
            wrap=True, horizontalalignment='center', fontsize=12)

plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Comparar con ecuación maestra para P_e(t)
plt.figure(figsize=(12, 6))
plt.plot(times, results['avg_pop'], '--', label='Simulación (Heterodina)')

# Ecuación maestra (Lindblad)
c_ops = [Ceg ,Cea, Cag]
rho0 = qt.ket2dm(Kg)   
result_me = qt.mesolve(H_sys, rho0, times, c_ops, [Aee, Aaa, Agg])
plt.plot(times, result_me.expect[0], 'k-', label='Lindblad', linewidth=2)

# Estilos de texto
plt.xlabel("Tiempo", fontsize=14)
plt.ylabel("Población $P_e(t)$", fontsize=14)
plt.title("Población del estado excitado", fontsize=16)

# Pie de figura con parámetros
plt.figtext(0.5, -0.1, f"Ω={Omega}, Δ={Delta}, γ={g}, t_max={t_max}, dt={dt}, Trayectorias={num_traj}",
            wrap=True, horizontalalignment='center', fontsize=12)

plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

```

```python
from scipy.signal import correlate
from scipy.fft import fft, fftfreq, fftshift
import matplotlib.pyplot as plt

num_traj, steps = trajectories_I.shape

# Autocorrelación promedio
autocorr_total = np.zeros(2*steps - 1, dtype=np.complex128)

for k in range(num_traj):
    I = trajectories_I[k, :]  # Fotocorriente compleja
    autocorr = correlate(I, I, mode='full')  # autocorrelación cruzada
    autocorr_total += autocorr

# Promedio
autocorr_avg = autocorr_total / num_traj

# Tiempo de retardo asociado
lags = np.arange(-steps + 1, steps) * dt
positive_lags = lags >= 0
tau = lags[positive_lags]
autocorr_pos = autocorr_avg[positive_lags] 
# Transformada de Fourier: S(ω) = ∫ dτ e^{iωτ} <I(t+τ) I(t)>
S_w = fftshift(fft(autocorr_pos))
freqs = fftshift(fftfreq(len(tau), dt))
```

```python
# Graficar el espectro
plt.figure(figsize=(8, 4))
plt.plot(freqs, np.real(S_w) , label="Espectro Normalizado")
plt.xlabel("Frecuencia [ω]")
plt.ylabel("Espectro S(ω)")
plt.title("Espectro desde fotocorriente simulada")
plt.grid(True)
plt.axvline(x=Omega/(2*np.pi), color='r', linestyle='--', label=f'ω=±{Omega/(2*np.pi):.2f}')
plt.axvline(x=-Omega/(2*np.pi), color='r', linestyle='--')
plt.xlim(-3, 3)
plt.legend()

# Parámetros de simulación en el pie de figura
sim_params = f"Ω={Omega}, Δ={Delta}, γ={g}, γd={gd}, γa={ga}, t_max={t_max}, dt={dt}, Trayectorias={num_traj}"
plt.figtext(0.5, -0.15, sim_params, wrap=True, horizontalalignment='center', fontsize=12)

plt.tight_layout()
plt.show()
```
