```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt

NF = 3
Kg = basis(NF, 2)  # Estado base
Ka = basis(NF, 1)  # Estado metastable
Ke = basis(NF, 0)  # Estado excitado

# Estado inicial
psi0 = Kg 

# Parmetros del sistema
Omega = 1  # Frecuencia de Rabi
Delta = 0    # Desintona
ga = 0.015 
gd = 0.05
g = 1
deltat = 25000  # Aumentar resolucin temporal para tiempos espaciados
ntraj = 5000

# Operadores de colapso para todas las transiciones relevantes
Ceg = np.sqrt(g) * (Kg * Ke.dag())  # Emisin de fotn detectable (excited to base)
Cea = np.sqrt(gd) * (Ka * Ke.dag())  # Transicin de excitado a metastable
Cag = np.sqrt(ga) * (Kg * Ka.dag())  # Transicin de metastable a base
collapse = [Ceg, Cea, Cag]

# Hamiltoniano
H = Delta * (Ke * Ke.dag()) + 0.5 * Omega * (Ke * Kg.dag() + Kg * Ke.dag())

# Lista de tiempos para la simulacin
tlist = np.linspace(0, 900, deltat)

# Listas para almacenar tiempos de espera clasificados
direct_times = []
metastable_times = []

# Simulacin Monte Carlo para medir tiempos de espera
for _ in range(ntraj):
    result = mcsolve(H, psi0, tlist, collapse, [], 1)
    traj = result.col_times[0]  # Extraer tiempos de colapso de esta trayectoria
    col_events = result.col_which[0]  # Identificar qu colapsos ocurrieron

    # Variable para registrar el primer tiempo de espera y su tipo
    last_time = None
    went_through_metastable = False  # Para saber si pas por el metastable

    for i, col_time in enumerate(traj):
        if col_events[i] == 0:  # Solo registramos tiempos entre colapsos de tipo 0
            if last_time is not None:
                # Si hubo un colapso de tipo 1 o 2 entre el primer y segundo colapso de tipo 0
                if went_through_metastable:
                    metastable_times.append(col_time - last_time)
                else:
                    direct_times.append(col_time - last_time)
                break  # Salir despus de registrar el primer tiempo entre colapsos de tipo 0

            # Actualizar el ltimo tiempo de colapso de tipo 0
            last_time = col_time
            went_through_metastable = False  # Reiniciar para la nueva medicin

        # Marcar si hubo un paso por el metastable entre dos colapsos de tipo 0
        elif col_events[i] in [1, 2]:
            went_through_metastable = True

# Mostrar resultados
print("Tiempos de espera (transicin directa):", direct_times)
print("Tiempos de espera (pasando por metastable):", metastable_times)

# Graficar los tiempos
plt.hist(direct_times, bins=50, alpha=0.7, label="Directo (excitado a base)")
plt.hist(metastable_times, bins=50, alpha=0.7, label="Pasando por metastable")
plt.xlabel("Tiempo")
plt.ylabel("Frecuencia")
plt.legend()
plt.title("Distribucin de Tiempos de Espera: Directo vs. Pasando por Metastable")
plt.show()
```
