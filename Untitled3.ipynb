```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt

NF = 3
Kg = basis(NF, 2)
Ka = basis(NF, 1)
Ke = basis(NF, 0)

# Estado inicial
psi0 = Kg 

# Definición de operadores de colapso y Hamiltoniano
Aaa = Ka * Ka.dag()
Aee = Ke * Ke.dag()
Agg = Kg * Kg.dag()
Aeg = Ke * Kg.dag()
Aag = Ka * Kg.dag()
Aea = Ke * Ka.dag()
Aae = Ka * Ke.dag()
Aga = Kg * Ka.dag()
Age = Kg * Ke.dag()

Omega = 3.5  # Frecuencia de Rabi
Delta = 0    # Desintonía
ga = 0.015 
gd = 0.05
g = 0.95
deltat = 10000
ntraj = 500

# Operadores de colapso para todas las transiciones relevantes
Ceg = np.sqrt(g) * Age  # Emisión de fotón detectable
Cea = np.sqrt(gd) * Aae  # Transición de excitado a metastable
Cag = np.sqrt(ga) * Aga  # Transición de metastable a base
collapse = [Ceg, Cea, Cag]

# Hamiltoniano
H = Delta * Aee + 0.5 * Omega * (Aeg + Age)

# Lista de tiempos para la simulación
tlist = np.linspace(0, 21, deltat)

# Listas para almacenar tiempos de espera
wait_times = []

# Simulación Monte Carlo para medir tiempos de espera
for _ in range(ntraj):
    result = mcsolve(H, psi0, tlist, collapse, [Aee, Agg, Aaa], 1)
    
    # Registrar el primer colapso de excitado a base
    for traj in result.col_times:
        if traj and len(traj) > 0:
            wait_times.append(traj[0])  # Primer tiempo de colapso de excitado a base
        else:
            # Si no hubo colapso en el tiempo simulado, asignar el tiempo máximo
            wait_times.append(tlist[-1])

# Análisis posterior: añadir tiempos oscuros
num_dark_times = int(0.05 * len(wait_times))  # 5% de tiempos oscuros
mean_dark_time = np.mean(wait_times) * 15  # Ajusta la escala para tiempos oscuros

# Generar tiempos oscuros adicionales
dark_times = np.random.exponential(scale=mean_dark_time, size=num_dark_times)

# Combinar tiempos originales y tiempos oscuros
combined_times = np.concatenate((wait_times, dark_times))

# Mostrar resultados
print("Tiempos de espera hasta el primer colapso (Aeg):", wait_times)
print("Tiempos oscuros añadidos:", dark_times)

# Graficar los tiempos
plt.hist(combined_times, bins=50, alpha=0.7, label="Tiempos combinados")
plt.xlabel("Tiempo")
plt.ylabel("Frecuencia")
plt.legend()
plt.title("Distribución de Tiempos de Espera con Tiempos Oscuros Añadidos")
plt.show()
```
